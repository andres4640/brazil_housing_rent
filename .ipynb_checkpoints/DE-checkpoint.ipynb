```python
intervals = pd.interval_range(start=0, end=100000, freq=5000)
total_intervals = pd.cut(df["total"], bins=intervals).value_counts().sort_index()*100

#Percentage distribution of total rent cost in intervals
total_intervals_per = pd.cut(df["total"], bins=intervals).value_counts(normalize=True).sort_index()*100
total_intervals_per
```

```python
print("Porcentage of houses with total cost of rent less than 30000 :", round(total_intervals_per[:5].sum(),2),"%")
```

We can see that 99.78% of the houses that are being rented have a total cost of rent less than 30000. Looks like values above that value could be outliers. Lets look for them.

```python
df.total.hist(grid=True,range=(400,30000), bins=12)
```

From the box plot we can see that there are a lot of outliers, we need to take them out so they don´t affect the analysis

```python
columns = []

df.boxplot(column="total", figsize=(10,5))
plt.ylim(0, 50000)
```

Actions for removing outliers:
1. Calculate intercuartil range
2. Multiply the intercuartil range by 1.5 and sum it with the highest quartile.
3. Clean the dataset and only accept rows with values lower that the operation made in step 3.

```python
round(df.total.describe(),2)
```

```python
from scipy import stats

iqr = stats.iqr(df["total"], interpolation = 'midpoint')
print("The intercuartile range is: ",iqr)
```

Cleaning the dataset:

```python
upper_quartile = df.total.describe()[6]
outlier_cleaner = upper_quartile + iqr*1.5
print("The values above ", outlier_cleaner, "will be considered outliers and will be erased")

df_clean = df[df["total"] <= outlier_cleaner]
```
