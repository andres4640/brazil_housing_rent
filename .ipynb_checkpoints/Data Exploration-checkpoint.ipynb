# Brazil house renting

Objective: 
- Show the most expensive cities.
- Total price distribution altogether and by city.
- See what are the trens in the houses in terms of area, rooms, bathrooms, parking spaces, floor, animal(acepted or not) and furniture, by city.
- Relationship between Price and different variables (area, rooms, bathrooms, parking space.. etc.)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("houses_to_rent_v2.csv")
df
```

```python
print("Dataframe description")
print("====================================================")
print(df.describe())
print("")
print("Dataframe DataTypes")
print("=====================================================")
print(df.info())
```

# Changing columns name

```python
df.rename(columns={"hoa (R$)":"hoa","rent amount (R$)":"rent_amount","property tax (R$)":"property_tax", "fire insurance (R$)":"fire_insurance","total (R$)":"total"}, inplace=True)
```

```python
house_columns = ["city","rooms", "bathroom", "parking spaces", "floor", "animal", "furniture"]

for i in house_columns:
    print("============ ",i, "plot =================")
    df[i].value_counts().sort_values().plot.bar(title=i)
    print(df[i].value_counts())
    plt.show()
    
plt.show()
```

# Analysis by city

```python
columns_analyse =  ["rooms", "bathroom", "parking spaces", "floor", "animal", "furniture"]
for i in columns_analyse:
    plt.figure(figsize=(12,6))
    plt.title(i)
    sns.countplot(x="city", hue=i, data=df)
    plt.show()
    
```

# Rent amount by city

We can see that the total rent in 98% of the houses in brazil cost less than 20000 reals.

```python
df.total.value_counts(normalize=True,bins=60).sort_index()*100
```

```python
df[df.total <= 30000]["total"].plot.hist(grid=True,xticks=np.arange(400,26401,2000), rot=30)
plt.xlim(400,26000)
plt.xlim(0,30000)
```

```python
df[df.total <= 20000].total.value_counts(bins=10)
```

Total cost plot is a right skewed distribution:

```python
df["total"].hist(grid=True,range=(400,36000), bins=12)
```

```python
cities_list = df.city.unique()

for i in cities_list:
    
    df[df.city == i]["total"].plot.kde(figsize=(12,5))
    
plt.xlim(0, 50000)
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.stripplot(x="city", y="total", data=df, jitter=True)
plt.ylim(0, 40000)
```

```python
plt.figure(figsize=(9,4))
sns.boxplot(x="city", y="total", data=df)
plt.ylim(0, 30000)
```

Summary statistics by city:

```python
df.groupby("city").agg(["count","min","max"])["total"]
```

```python
cities_list = df.city.unique()

for i in cities_list:
    df[df.city == i]["total"].plot.hist()
    plt.show()
```

# Variables correlation

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True, square=True)
plt.ylim(0,9)
```
